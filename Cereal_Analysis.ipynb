```python
import pandas as pd 
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cereal.csv')
```

```python
print(df.head(10))
```

```python
df.info()
```

```python
df.mfr.replace(to_replace=['A', 'G', 'K', 'N', 'P', 'Q', 'R'], value=['American Home Food Products','General Mills', 'Kellogs', 'Nabisco', 'Post', 'Quaker Oats', 'Ralston Purina'], inplace = True)
```

```python
df.head(10)
```

```python
#Any Correlations? 
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
#What are the top rated cereals? 

df.sort_values(by = 'rating', ascending = False).head(10)
top_cereals = df.sort_values(by = 'rating', ascending = False).head(5)
#print(top_cereals)

#Make a bar chart
top_cereals.plot(kind="bar", x = 'name', y = 'rating')
```

```python
df.sort_values(by = 'protein', ascending = False).head(5)
top_protein = df.sort_values(by = 'protein', ascending = False).head(5)

top_protein.plot(kind="bar", x = 'name', y = 'protein')
```

```python
df.sort_values(by = 'potass', ascending = False).head(10)
top_potass = df.sort_values(by = 'potass', ascending = False).head(10)

top_potass.plot(kind='bar', x = 'name', y = 'potass')
```

```python
#How many from each manufacturer? 
mfrs = df.mfr.value_counts()
mfrs.plot(kind='bar', x = 'name', y = 'mfrs')
```

```python
#Pie chart of manufacturers and number of cereals in this list 
mfrs = df.mfr.value_counts()
mfrs.plot(kind='pie')
```

```python
#Pie chart of cold and hot 
cold_hot = df.type.value_counts()
print(cold_hot)
```

```python
#Calories top 5 high and top 5 low 
calories_high = df.sort_values(by = 'calories', ascending = False).head(5)

calories_high.plot(kind = 'bar', x = 'name', y = 'calories')

calories_low = df.sort_values(by = 'calories', ascending = True).head(5)

calories_low.plot(kind = 'bar', x = 'name', y = 'calories')
```
